## from "2.21 Adding Text"

```python
pic = Picture(scale=3, line_cap='round')
pic.tikzstyle('axes')
pic.tikzstyle('important line', very_thick=True)

# Colors
pic.colorlet('anglecolor', 'green!50!black')
pic.colorlet('sincolor', 'red')
pic.colorlet('coscolor', 'blue')

# The graphic
pic.draw((-1.4, -1.4), grid((1.4, 1.4), step='0.5cm'),
    help_lines=True, very_thin=True)

pic.draw((0, 0), circle(radius=1))

scope = pic.add_scope(axes=True)
scope.draw((-1.5, 0), lineto((1.5, 0)), node('$x$', right=True), coordinate(name='x axis'),
    opt='->')
scope.draw((0, -1.5), lineto((0, 1.5)), node('$y$', above=True), coordinate(name='y axis'),
    opt='->')

for x in [-1, -0.5, 0, 0.5, 1]:
    scope.draw('(0pt,1pt)', lineto('(0pt,-1pt)'), node(f'${x}$', below=True, fill='white'),
        xshift=f'{x} cm')

for y in [-1, -0.5, 0, 0.5, 1]:
    scope.draw('(1pt,0pt)', lineto('(-1pt,0pt)'), node(f'${y}$', left=True, fill='white'),
        yshift=f'{y} cm')

pic.path((0, 0), lineto('(3mm,0mm)'), arc(start_angle=0, end_angle=30, radius='3mm'), lineto(cycle()),
    fill='green!20', draw='anglecolor')
pic.draw(polar(15, '2mm'), node(r'$\alpha$', anglecolor=True))

pic.draw(polar(30, '1cm'), lineto(vertical(polar(30, '1cm'), 'x axis')),
    node(r'$\sin \alpha$', left='1pt', fill='white', pos=0.5),
    important_line=True, sincolor=True)

pic.draw(vertical(polar(30, '1cm'), 'x axis'), lineto((0,0)),
    node(r'$\cos \alpha$', below='2pt', fill='white', pos=0.5),
    important_line=True, coscolor=True)

pic.usetikzlibrary('intersections')
pic.path((1, 0), lineto((1, 1)), name_path='upward line')
pic.path((0, 0), lineto(polar(30, 1.5)), name_path='sloped line')
pic.draw(name_intersections='{of=upward line and sloped line, by=t}')

pic.draw((1, 0), lineto('(t)'), node(
            r"""$
            \displaystyle
            \tan \alpha
            \color{black} = \frac{
                \color{red} \sin \alpha}{
                \color{blue} \cos \alpha}
            $""",
            right='1pt', fill='white', pos=0.5),
        orange=True, very_thick=True)

pic.draw((0, 0), lineto('(t)'))

pic.demo()
```

***

A coordinate (in path specifications as well as as arguments elsewhere) is a string, or a `tuple` or 1d-`ndarray` with 2 or 3 elements.

- Strings can be used e.g. to provide coordinates in TikZ' `canvas` coordinate system. Coordinate-specifying strings start with '(' and end with ')'.
- Elements of `tuple`s can be numbers or strings.
  - If all elements are numeric, it represents coordinates in TikZ' `xyz` coordinate system and is converted into a 1d-`ndarray`,
  - if all are strings it represents coordinates in TikZ' `canvas` coordinate system, and is converted into a simple string including parentheses.
  - otherwise it represents a mixed `xyz`/`canvas` coordinate as described in 13.2.1 and is left as a tuple.
 - `ndarray`s must be numeric. 

A sequence of coordinates is a `list` of coordinates as described above, or a numeric 2d-`ndarray` with 2 or 3 columns, representing `xyz` coordinates. If a list contains only numeric 1d-`ndarray`s (after conversion) with the same number of elements, it is converted into a 2d-`ndarray` itself.

Redesign:

- `elements` of an environment object (`Scope` and descendants) can be `Action` objects or strings. When the `code` method is called on the environment object, it returns a multiline string composed of its elements. String elements are written as themselves into one line each. For action objects their `code` method is called, which returns a string to be written.

- `Action` objects are appended to an environment's `elements` via environment methods (`path` and friends). These methods accept an arbitrary number of non-keyword arguments (`*spec`), as well as options for the action (`opt=` and `**kwoptions`) which are used to construct the path specification. The path specification elements can be strings (stored as such), coordinates, lists of coordinates, and `Operation` elements. (Lists of) coordinates are converted into (one or more) 'move-to' `Action` objects.

- Coordinates stored in `Action` objects that are partially or completely in the `xyz` coordinate system may be transformed before written to TikZ code, if they occur within an `Axis` where they are interpreted as data.

- Strings are allowed as `elements` of an environment, part of a path specification, and as coordinates in order to make TikZ features accessible that are not explicitly modeled.
